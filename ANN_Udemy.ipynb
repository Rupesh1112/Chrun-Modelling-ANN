The dataset is bank customer data and we have to predict whether customer will leave the bank or not.
Dataset: Churn_Modelling 
Number of observations: 10000    

```python
# Classification template

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from sklearn import preprocessing
```

```python
#PART 1: DATA PREPROCESSING
#Import dataset 
#Set X equal to independent variables in dataset and y equals to final column which is prediction 
dataset = pd.read_csv('Churn_Modelling.csv')
#dataset.shape()
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
print(X)
print(y)
```

```python
#Encoding the categorical data (here gender and country of the customer)
#Label Encoding of "Gender Coloumn" 
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X [:,2] = le.fit_transform(X[:,2])  #2th coloumn contains gender information
print(X)
```

Now we have to encode the geogrophy coloumn. But we cannot encode it simply as 1 and 0 as it may loss the information while 
making the equation. Thus we need to use "One Hot Encoder" here.
Here we have encoded the countries as
France : 1 0 0 
Spain  : 0 0 1
Germany: 0 1 0

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[1])],remainder = 'passthrough') #[1] means column 1 == Gepgraphy
X = np.array(ct.fit_transform(X))
print(X)
```

```python
#Split the Dataset into Training and Testing Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Feature Scaling is very very important in deep learning.
#It is fundamental step in ANN or Deep Learning
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
```

```python
#Part 2: BUILDING THE ANN
#2.1 : Initializing the ANN
import keras
ann = tf.keras.models.Sequential()
```

```python
#2.2 Adding Input and the first hidden layer
#We will add fully connected hidden layer and input layer get automatically created
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

```python
#2.3 Adding second hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

```python
#2.4 Adding output layer
#Since y is having two classes, thus number of neurons in output layer is 1
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
#PART 3: TRAINING THE ANN
#3.1 Compile the ANN
ann.compile(optimizer ='Adam' ,loss ='binary_crossentropy' ,metrics =['accuracy']) #Plz study other types of parameters
```

```python
#3.2 Training the ANN on training set
ann.fit(X_train, y_train, batch_size = 32,epochs = 200 )#batch size generally used is 32 # study epoch
#we need number of epochs to learn the ANN over time and in more accurate manner
#accuarcy increases as epochs
```

```python
#4.2 Prediciting the test set results
y_pred = ann.predict(X_test)
y_pred = (y_pred>0.5)
print(np.concatenate(y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)))
```

```python
#4.3 Calculate confusion matrix and accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
